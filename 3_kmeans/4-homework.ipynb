```python
path = ("circle.csv")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv(path)
data = np.array(data)
```

```python
print(data)
print(data)
```

```python
x1 = []
y1 = []
x2 = []
y2 = []
for item in data:
    if(item[2]==0):
        x1.append(item[0])
        y1.append(item[1])
    else:
        x2.append(item[0])
        y2.append(item[1])
print(len(x1))
print(len(y1))
print(len(x2))
print(len(y2))
```

```python
plt.scatter(x1, y1, marker='*', s=100)
plt.scatter(x2, y2, marker='<', s=100)
plt.savefig('data_cluster.pdf')
plt.show()
```
